```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# === LOAD DATASET ===
df = pd.read_csv("heart_failure_clinical_records_dataset.csv")
print("Dataset berhasil dibaca!")
print(df.head())
```

```python
# Memisahkan fitur dan target
X = df.drop("DEATH_EVENT", axis=1).values
y = df["DEATH_EVENT"].values
```

```python
# Split manual 80% train, 20% test
split = int(len(X) * 0.8)
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]
```

```python
# Normalisasi manual
mean = X_train.mean(axis=0)
std = X_train.std(axis=0)
```

```python
X_train = (X_train - mean) / std
X_test  = (X_test - mean) / std
```

```python
class SimpleTree:
    def __init__(self):
        self.feature_index = None
        self.threshold = None
        self.value = None

    def fit(self, X, y):
        corrs = []
        for i in range(X.shape[1]):
            try:
                corr = abs(np.corrcoef(X[:, i], y)[0, 1])
            except:
                corr = 0
            corrs.append(corr)

        self.feature_index = int(np.argmax(corrs))
        self.threshold = np.median(X[:, self.feature_index])
        self.value = 1 if y.mean() > 0.5 else 0

    def predict(self, X):
        return np.where(X[:, self.feature_index] > self.threshold, self.value, 1 - self.value)


class SimpleRandomForest:
    def __init__(self, n_estimators=7):
        self.trees = [SimpleTree() for _ in range(n_estimators)]

    def fit(self, X, y):
        for tree in self.trees:
            idx = np.random.choice(len(X), len(X), replace=True)
            tree.fit(X[idx], y[idx])

    def predict(self, X):
        preds = np.array([tree.predict(X) for tree in self.trees])
        return (preds.mean(axis=0) >= 0.5).astype(int)
```

```python
print("Melatih model Simple Random Forest...")

rf = SimpleRandomForest(n_estimators=7)
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)

print("Training selesai.")
```

```python
def confusion_matrix_manual(y_true, y_pred):
    cm = np.zeros((2, 2), dtype=int)
    for yt, yp in zip(y_true, y_pred):
        cm[yt][yp] += 1
    return cm


def tampilkan_evaluasi(nama, y_true, y_pred):
    print(f"\n=== HASIL EVALUASI {nama} ===")

    # Akurasi
    acc = (y_true == y_pred).mean()
    print(f"Akurasi: {acc:.2%}")

    # Confusion Matrix
    cm = confusion_matrix_manual(y_true, y_pred)
    print("\nConfusion Matrix:")
    print(cm)

    # Plot CM
    plt.figure(figsize=(5, 4))
    plt.imshow(cm)
    plt.title(f"Confusion Matrix - {nama}")
    plt.xlabel("Prediksi")
    plt.ylabel("Aktual")

    for i in range(2):
        for j in range(2):
            plt.text(j, i, cm[i][j], ha="center", va="center")

    plt.colorbar()
    plt.show()

tampilkan_evaluasi("Simple Random Forest", y_test, rf_pred)
```

```python
def plot_importance_manual(X, y):
    importances = []
    for i in range(X.shape[1]):
        try:
            imp = abs(np.corrcoef(X[:, i], y)[0, 1])
        except:
            imp = 0
        importances.append(imp)

    fitur = df.drop("DEATH_EVENT", axis=1).columns
    idx = np.argsort(importances)[::-1]

    plt.figure(figsize=(10, 5))
    plt.bar(fitur[idx], np.array(importances)[idx])
    plt.xticks(rotation=45)
    plt.title("Feature Importance (Manual)")
    plt.tight_layout()
    plt.show()

plot_importance_manual(X, y)
```
